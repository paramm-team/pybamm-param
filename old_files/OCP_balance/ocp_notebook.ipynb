```python
import ocp_balance
import pandas as pd

#Loading files here
anode_half_cell_data = pd.read_csv("anode_OCP_2.csv", header=None)
cathode_half_cell_data = pd.read_csv("cathode_OCP_2.csv", header=None)
anode_three_electrode_data = pd.read_csv("anode_OCP_3.csv", header=None)
cathode_three_electrode_data = pd.read_csv("cathode_OCP_3.csv", header=None)

#These data then supplied to the package (The order is important!)
ocp = ocp_balance.ocp(anode_half_cell_data,
                      cathode_half_cell_data,
                      anode_three_electrode_data,
                      cathode_three_electrode_data) 
```

```python
#Optimized shifting values are found by Least Squares Method
ocp.ocp_optimize()
```

```python
'''Relationship between Capacity and Stoichiometry is:
Stoichiometry = (a_anode)*(Capacity)+b_anode

Please provide a_anode and b_anode'''

ocp.anode_plot(a_anode=0.1974, b_anode = 0.0218)
```

```python
'''Relationship between Capacity and Stoichiometry is:
Stoichiometry = (a_cathode)*(Capacity)+b_cathode

Please provide a_cathode and b_cathode'''

ocp.cathode_plot(a_cathode=-0.1451, b_cathode = 0.9660)
```
